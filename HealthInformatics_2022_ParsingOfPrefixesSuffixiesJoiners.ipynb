```python
dfs = []
for table in tables:
  if table.findParent("table") is None:
    df=pd.read_html(str(table))
    # convert list to dataframe
    df=pd.DataFrame(df[0])
    if 'Affix' in df:
      dfs.append(df)
      display(df.head(n=10))                # display first few elements of the table
```

# Preprocessing the data
The Wikipedia table has a single column for affixes, that is, it does not give prefixes and suffixes separately. 
We are distinguishing them by looking for hyphens('-').


```python
# dfs[0]['Affix'].tolist()
import re

prefixes=""
suffixes=""
conns=""
for i in range(len(dfs)):
  l=dfs[i]['Affix'].tolist()
  prefs = [af for af in l if af[-1]=='-']   # if hyphen is present at the end of the affix, it is a prefix
  sufs = [af for af in l if af[0]=='-']     # if hyphen is present at the beginning of the affix, it is a suffix
  prefixes+=str(' '.join(prefs))
  suffixes+=str(' '.join(sufs))

chars=[',','-']
for c in chars:
  prefixes=prefixes.replace(c,'')
  suffixes=suffixes.replace(c,'')

prefixes=re.sub('\(.*?\)','',prefixes)        # the characters present between open braces ('(') and close braces (')') are connectors
suffixes=re.sub('\(.*?\)','',suffixes)
```

```python
# suffixes.find('algia')
prefixes.find('my')
```

```python
extra_suffixes = ['porosis', 'lysis', 'some']
suffixes = suffixes + ' ' + ' '.join(extra_suffixes)
prefixes = prefixes + ' ' + 'lact'
```

```python
list_of_prefixes = "ab ante anti bi bin brady de di dia dys endo epi eu hemi hyper hypo inter intra macro meta micro multi neo pachy pan para per peri poly post pre pro sub supra sym syn trans tri tachy uni"
toks = list_of_prefixes.split(" ")
for t in toks:
  prefixes = prefixes.replace(' '+t+' ',' ')
```

```python
endOfString = StringEnd()
prefix = oneOf(list_of_prefixes)
root = oneOf(prefixes)
suffix = oneOf(suffixes)
connector = oneOf("o i si ato at e to")
```

# Defining the Regex

In order to parse the terminology, we need a regex which can capture the structure of the word. 
**We are following a simple regex which parses: prefix + connector + root + connector + suffix with the exception that the connector 'o' is not required if the suffix begins with a vowel**


```python
s1 = (root('root')+connector('connector')+suffix('suffix'))+endOfString
s2 = (prefix('prefix')+connector('connector1')+root('root')+connector('connector2')+suffix('suffix'))+endOfString
s3 = (root('root1')+connector('connector1')+root('root2')+connector('connector2')+suffix('suffix'))+endOfString
s4 = (root('root')+suffix('suffix'))+endOfString
s5 = (prefix('prefix')+Optional(connector('connector1'))+root('root')+Optional(connector('connector2')))+endOfString #Why no suffix in s5? # tachycardia, otherwise it is s2
s6 = (prefix('prefix')+suffix('suffix'))+endOfString
s7 = (root('root1')+connector('connector1')+root('root2')+suffix('suffix'))+endOfString             # for gastroenteritis type words
s8 = (prefix('prefix')+root('root'))+endOfString

print(connector)
word = (s5 | s1 | s2 | s3 | s4 | s6 | s7 | s8)
 
wordlist = ['arthroscope', 'arthritis', 'dialysis', 'osteoarthritis', 'dysarthr'] #'dysarthria', 'tachycardia', 
#wordlist = ['gastroenteritis', 'cardiomyopathy', 'arthritis', 'arthroscope', 'myopathy', 'glycolysis', 'bronchitis', 'dialysis', 'osteoporosis', 'diarrhea']
for wd in wordlist:
    print(wd)
    res = word.parse_string(wd)
    print(res.dump())
    # print(res.prefix)
    # print(res.root)
    # print(res.suffix)
    print()
```

# Part 2 : Clustering and Disambiguation

```python
import numpy as np
import torch
import torch.nn as nn
from transformers import AutoTokenizer, AutoModel

# Use last four layers by default
layers = [-6, -5, -4, -3, -2, -1]
tokenizer = AutoTokenizer.from_pretrained("bert-base-cased")
model = AutoModel.from_pretrained("bert-base-cased", output_hidden_states=True)
 

def get_word_idx(sent: str, word: str):
    return sent.split(" ").index(word)


def get_hidden_states(encoded, token_ids_word, model, layers):
    """Push input IDs through model. Stack and sum `layers` (last four by default).
      Select only those subword token outputs that belong to our word of interest
      and average them."""
    with torch.no_grad():
        output = model(**encoded)

    # Get all hidden states
    states = output.hidden_states
    # Stack and sum all requested layers
    output = torch.stack([states[i] for i in layers]).sum(0).squeeze()
    # Only select the tokens that constitute the requested word
    word_tokens_output = output[token_ids_word]

    return word_tokens_output.mean(dim=0)


def get_word_vector(sent, idx=0, tokenizer=tokenizer, model=model, layers=layers):
  """Get a word vector by first tokenizing the input sentence, getting all token idxs
    that make up the word of interest, and then `get_hidden_states`."""
  encoded = tokenizer.encode_plus(sent, return_tensors="pt")
  # get all token idxs that belong to the word of interest
  token_ids_word = np.where(np.array(encoded.word_ids()) == idx)

  return get_hidden_states(encoded, token_ids_word, model, layers)
 

def similarity(word1, word2, metric = nn.CosineSimilarity(dim=0, eps=1e-08)):
  emb1, emb2 = get_word_vector(word1), get_word_vector(word2)
  return metric(emb1,emb2)


```

```python
print(similarity('super','ultra'))
print(similarity('super','sub'))
```

```python
# # !pip3 install sentence-transformers
sentences = [
    'retro', 'sub', 'super', 'tachy', 'trans', 'ultra'
]
tokenizer = AutoTokenizer.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')
model = AutoModel.from_pretrained('sentence-transformers/bert-base-nli-mean-tokens')

# from sentence_transformers import SentenceTransformer

# model = SentenceTransformer('bert-base-nli-mean-tokens')

# sentence_embeddings = model.encode(sentences)
```

```python
tokens = {'input_ids': [], 'attention_mask': []}

for sentence in sentences:
    # encode each sentence and append to dictionary
    new_tokens = tokenizer.encode_plus(sentence, max_length=128,
                                       truncation=True, padding='max_length',
                                       return_tensors='pt')
    tokens['input_ids'].append(new_tokens['input_ids'][0])
    tokens['attention_mask'].append(new_tokens['attention_mask'][0])

# reformat list of tensors into single tensor
tokens['input_ids'] = torch.stack(tokens['input_ids'])
tokens['attention_mask'] = torch.stack(tokens['attention_mask'])
outputs = model(**tokens)
embeddings = outputs.last_hidden_state
attention_mask = tokens['attention_mask']
mask = attention_mask.unsqueeze(-1).expand(embeddings.size()).float()
masked_embeddings = embeddings * mask
summed = torch.sum(masked_embeddings, 1)
summed_mask = torch.clamp(mask.sum(1), min=1e-9)
mean_pooled = summed / summed_mask
from sklearn.metrics.pairwise import cosine_similarity

# convert from PyTorch tensor to numpy array
mean_pooled = mean_pooled.detach().numpy()
```

```python
scores = cosine_similarity(
    [mean_pooled[0]],
    mean_pooled
)[0]
```
